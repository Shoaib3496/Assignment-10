1. What is Simple Linear Regression
 - Simple Linear Regression (SLR) is one of the most basic and widely used techniques in statistical modeling and machine learning. It helps us understand and predict the relationship between two variables — one independent variable (X) and one dependent variable (Y).
 - Simple Linear Regression is a method that models the relationship between a dependent variable (Y) and an independent variable (X) by fitting a straight line through the data points.
 - That line is called the regression line, and it shows how Y changes as X changes.
 - The mathematical form is: Y = β0​+β1​X+ε, where
 - Y → Dependent variable (the one you want to predict)
 - X → Independent variable (the one you use for prediction)
 - β₀ (beta zero) → Intercept (value of Y when X = 0)
 - β₁ (beta one) → Slope (how much Y changes for one unit change in X)
 - ε (epsilon) → Error term (difference between the actual and predicted Y)
 - Assumptions of Simple Linear Regression:
 - Linearity → The relationship between X and Y is linear.
 - Independence → Observations are independent of each other.
 - Homoscedasticity → Constant variance of errors across all values of X.
 - Normality of errors → The residuals (errors) are normally distributed.
 - Applications:
 - Predicting sales based on advertising spend
 - Estimating house price based on square footage
 - Forecasting temperature from CO₂ concentration
 - Predicting marks from study hours

2. What are the key assumptions of Simple Linear Regression
 - 1.Linearity
 - The relationship between the independent variable (X) and the dependent variable (Y) must be linear.
 - In other words, a straight line should best describe how Y changes as X changes.
 - Example:
 - Marks increase roughly in proportion to study hours (linear).
 - Marks increase exponentially with study hours (non-linear).
 - How to check:
 - Plot a scatter plot of X vs Y — it should look roughly like a straight line.
 - Or check residual vs fitted value plot — residuals should be randomly scattered (not curved).

 - 2.Independence of Errors
 - The residuals (errors) should be independent of each other.
 - No correlation should exist between consecutive errors.
 - Example:
 - If you’re predicting stock prices daily, today’s prediction error should not depend on yesterday’s error.
 - How to check:
 - Use Durbin-Watson test — value close to 2 indicates independence.
 - Plot residuals over time — they should not form a pattern.

 - 3.Homoscedasticity (Constant Variance of Errors)
 -The residuals should have constant variance across all levels of X.
 - If variance increases or decreases with X, it’s called heteroscedasticity, and it violates this assumption.
 - Example:
 - Errors are evenly spread for all values of X.
 - Errors spread widely for large X values but are tight for small X values.
 - How to check:
 - Plot residuals vs predicted values.
 - ➤ If the spread looks even — good.
 - ➤ If you see a funnel or cone shape — it’s a problem

 - 4.Normality of Errors
 - The residuals (errors) should be normally distributed (bell-shaped curve).
 - This helps in making valid confidence intervals and hypothesis tests.
 - How to check:
 - Draw a histogram or Q–Q plot of residuals.
 - Perform Shapiro–Wilk test or Kolmogorov–Smirnov test.

 - 5.No (or minimal) Multicollinearity
 - In Simple Linear Regression, there is only one independent variable, so multicollinearity doesn’t apply.
 - However, in Multiple Linear Regression, independent variables shouldn’t be highly correlated with each other.

 - 6.No Measurement Errors in X
 - The independent variable (X) should be measured accurately.
 - If X has large measurement errors, the regression estimates (slope and intercept) will be biased.

3. What is heteroscedasticity, and why is it important to address in regression
models?
 - Heteroscedasticity (pronounced het-er-oh-skeh-DAST-ick-city) means that the variance of the error terms (residuals) is not constant across all levels of the independent variable(s).
 - In simpler words, The spread of errors changes as the value of X (predictor) changes.
 - Imagine plotting your regression model’s residuals vs predicted values:
 - If the residuals are evenly scattered around zero → Homoscedasticity (good)
 - If the residuals form a pattern or cone shape (funnel) → Heteroscedasticity (bad)
 - Example: In a model predicting food expenditure, people with higher incomes might have a wider range of spending compared to people with lower incomes, who have a more limited range. This unequal variance in the error term is heteroscedasticity
 - Importance:
 - Violation of OLS assumption: OLS regression assumes that the variance of the error term is constant (homoscedasticity). Heteroscedasticity violates this assumption.
 - Invalidates statistical inference: When heteroscedasticity is present, the standard errors of the coefficients are biased. This means that the p-values from hypothesis tests and the confidence intervals are no longer reliable and can lead to incorrect conclusions about the significance of your variables.
 - Inefficient estimates: The coefficient estimates are still unbiased, but they become less precise and efficient. This means they may be farther from the true population values.
 - How to address it
 - Weighted Least Squares (WLS): Use WLS to give less weight to observations with higher variance and more weight to observations with lower variance, minimizing the sum of weighted squared residuals.
 - Data transformation: Transform the dependent variable (or other variables) to stabilize the variance. Be aware that this can make the results harder to interpret in the original units.
 - Model re-specification: Re-examine the model to ensure it is correctly specified, as an incorrect functional form or missing variables can cause heteroscedasticity.
 - Use robust standard errors: Some statistical packages can calculate robust standard errors that are valid even in the presence of heteroscedasticity.

4. What is Multiple Linear Regression
 - Multiple linear regression is a statistical technique that models the relationship between a single dependent variable and two or more independent variables. It's an extension of simple linear regression that uses a linear equation to estimate how changes in multiple predictor variables affect an outcome variable, allowing for more precise predictions than using a single predictor.
 - How it works:
 - Model structure: It uses a linear equation to estimate the relationship between variables. The model can be visualized as a plane or hyperplane in higher dimensions.
 - Independent variables: Two or more predictor variables that are used to predict the outcome.
 - Coefficients: The model produces coefficients (B) for each independent variable. These indicate how much the dependent variable is expected to change for a one-unit increase in that independent variable, assuming all other independent variables are held constant
 - Equation: Y = β0​+β1​X+β2X2​+...+βn​Xn
 - Applications
 - Forecasting: Making predictions based on historical data.
 - Understanding relationships: Determining which factors have the greatest influence on an outcome.
 - Testing theories: Examining and explaining relationships between financial or social variables.
 - Examples:
 - Predicting a house's price based on its square footage, number of bedrooms, and location.
 - Determining how a student's test score is affected by hours studied, previous grades, and class attendance.
 - Key metrics R-squared ((R^{2})): Indicates the proportion of the variance in the dependent variable that is predictable from the independent variables. - - - Adjusted R-squared: A modified version of R-squared that accounts for the number of predictors in the model, which is useful to avoid overestimating the model's fit when many independent variables are used.
 - F-test: A statistical test that checks the overall significance of the model by determining if at least one of the independent variables has a statistically significant relationship with the dependent variable.
 - P-value: Used to determine the statistical significance of individual coefficients. A low p-value (typically < 0.05) suggests that the corresponding independent variable has a significant impact on the dependent variable.

5. What is polynomial regression, and how does it differ from linear
regression
 - Polynomial Regression:
 - Polynomial Regression is a type of regression analysis that models the relationship between the independent variable (X) and the dependent variable (Y) as an nth-degree polynomial.
 - It's still a linear model in terms of coefficients, but it can capture nonlinear relationships between X and Y.
 - For degree n, the model is: Y = β0​+β1​X+β2X^2​+...+βn​X^n+ε
 - where
 - Y → Dependent variable
 - X → Independent variable
 - β₀, β₁, ..., βₙ → Coefficients to be learned
 - ε → Error term
 - A type of regression analysis that models the relationship between a dependent variable and one or more independent variables as an n-th degree polynomial.
 - Used when the data points show a curvilinear pattern that a straight line cannot accurately represent.
 - It can capture more complex, non-linear patterns in the data.
 - Prone to overfitting, where the model becomes too complex and fits the noise in the data, especially with a high-degree polynomial. Choosing the right degree is crucial.
 - Linear regression:
 - A statistical method that models the relationship between a dependent variable and one or more independent variables as a straight line.
 - Equation: Y = β0​+β1​X+ε
 - Used when the data points exhibit a linear relationship, or a straight line is sufficient to model the pattern.
 - Advantage: Simpler to understand and interpret than polynomial regression.
 - Risk: Can result in underfitting if the relationship between variables is actually non-linear but a straight line is forced upon it.

6. Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)

print("Model Intercept (b0):", model.intercept_)
print("Model Slope (b1):", model.coef_[0])

plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Actual Data Points')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Simple Linear Regression Fit')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.legend()
plt.grid(True)
plt.show()
```

7. Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

X = df[['Area', 'Rooms']]
Y = df['Price']

model = LinearRegression()
model.fit(X, Y)

print("Model Intercept:", model.intercept_)
print("Model Coefficients:", model.coef_)

X_vif = pd.DataFrame(data)
X_vif = X_vif[['Area', 'Rooms']]
vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(len(X_vif.columns))]

print("\nMulticollinearity check (VIF):")
print(vif_data)
```

8. Implement polynomial regression on the following data: ● X = [1, 2, 3, 4, 5] ● Y = [2.2, 4.8, 7.5, 11.2, 14.7] Fit a 2nd-degree polynomial and plot the resulting curve. (Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, Y)

X_smooth = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_smooth_poly = poly.transform(X_smooth)
Y_pred = model.predict(X_smooth_poly)

print("Model Intercept:", model.intercept_)
print("Model Coefficients:", model.coef_)

plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Actual Data Points')
plt.plot(X_smooth, Y_pred, color='red', linewidth=2, label='2nd-Degree Polynomial Curve')
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.legend()
plt.grid(True)
plt.show()
```

9.  Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)
residuals = Y - Y_pred

print("Model Intercept:", model.intercept_)
print("Model Slope:", model.coef_[0])

plt.figure(figsize=(8, 6))
plt.scatter(Y_pred, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals Plot')
plt.xlabel('Predicted Y Values')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```

10. Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
 - However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model.
 - Multicollinearity occurs when independent variables are highly correlated with each other, making it difficult for the model to estimate the unique effect of each variable on the price.
 - Steps for Mitigation:
 - 1.Diagnosis (VIF): I would first quantify the extent of the issue using the Variance Inflation Factor (VIF). Features with a VIF score greater than 5 or 10 are typically considered to have significant multicollinearity.
 - 2.Feature Selection:
 - Dropping Redundant Features: If two features like Area and Number of rooms are re highly correlated (larger houses usually have more rooms), I might choose to keep only the one that is most strongly correlated with the target variable or the one that is more intuitively significant to stakeholders.
 - Domain Expertise: I would use my knowledge of the real estate market to make informed decisions about which features are essential for prediction.
 - 3.Feature Engineering:
 - Creating Interaction Terms: Instead of removing features, I could create a new, combined feature, such as "Area per Room".
 - 4.Regularization Techniques:
Ridge Regression or Lasso Regression: These techniques add a penalty to the model's coefficients, which helps manage the impact of correlated features and can prevent the coefficients from becoming excessively large and unstable. Lasso also performs feature selection by shrinking some coefficients to zero.
 - 5.Principal Component Analysis (PCA): As a last resort for severe cases, PCA can transform the correlated features into a new set of uncorrelated components (principal components) that capture the majority of the variance in the data.
 - Addressing Heteroscedasticity:
 - Heteroscedasticity means that the variance of the error terms (residuals) is not constant across all levels of the independent variables. For example, my model might make small, consistent errors for low-priced homes but large, variable errors for high-priced luxury homes. This violates a key assumption of ordinary least squares (OLS) regression and leads to unreliable standard errors and p-values.
 - Steps for Mitigation:
 - 1.Diagnosis (Residuals Plot): I would examine a residuals plot, plotting the residuals against the predicted values. A "fanning" or conical shape indicates heteroscedasticity, whereas a random, uniform scatter indicates homoscedasticity (constant variance).
 - 2. Data Transformation:
 - Log Transformation: Applying a log transformation to the dependent variable (Price) is a very common and effective technique in real estate. It often stabilizes the variance and also helps to normalize the typically right-skewed distribution of house prices.
 - Square Root or Cube Root Transformations: These can also be effective for managing variance.
 - Weighted Least Squares (WLS) Regression: If transformations are not ideal, WLS assigns different weights to different observations. Data points associated with higher variance (e.g., expensive homes with large errors) are given lower weights, effectively giving more importance to more reliable data points.
 - Robust Standard Errors: I could use robust standard errors (also known as Huber-White standard errors) during model estimation. This approach does not correct the heteroscedasticity itself but provides correct and reliable standard errors and confidence intervals despite the issue.
 - By systematically applying these techniques, I can build a more stable, interpretable, and statistically robust model for predicting house prices.
